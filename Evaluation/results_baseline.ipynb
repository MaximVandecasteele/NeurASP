```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# RGB baseline learning curve

```python
file_path = '/Users/maximvandecasteele/PycharmProjects/NeurASP/Evaluation/training_run_baseline/csv_rgb/rgb_run1.csv'
file_path2 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/Evaluation/training_run_baseline/csv_rgb/rgb_run2.csv'
file_path3 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/Evaluation/training_run_baseline/csv_rgb/rgb_run3.csv'
file_path4 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/Evaluation/training_run_baseline/csv_rgb/rgb_run4.csv'
file_path5 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/Evaluation/training_run_baseline/csv_rgb/rgb_run5.csv'
```

```python
df1 = pd.read_csv(file_path).drop('Wall time', axis=1)
df2 = pd.read_csv(file_path2).drop('Wall time', axis=1)
df3 = pd.read_csv(file_path3).drop('Wall time', axis=1)
df4 = pd.read_csv(file_path4).drop('Wall time', axis=1)
df5 = pd.read_csv(file_path5).drop('Wall time', axis=1)
```

```python
df1['episode']
```

```python
value_columns = [df['reward'] for df in [df1, df2, df3, df4, df5]]
concatenated_values = pd.concat(value_columns, axis=1)
result_df = pd.DataFrame(concatenated_values.mean(axis=1), columns=['mean_value'])
result_df_rgb_training = pd.concat([df1['episode'], result_df], axis=1)
```

```python
result_df_rgb_training
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Plot the average reward per model
plt.figure(figsize=(10, 6))  # Adjust size as needed
sns.lineplot(x='episode', y='mean_value', data=result_df_rgb_training, sort=False)
plt.xlabel('Episode')
plt.ylabel('Average Total Reward')
plt.title('Average Total reward per Episode')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
result_df_rgb_training['ewm'] = result_df_rgb_training.ewm(alpha=0.1).mean()['mean_value']
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Plot the average reward per model
plt.figure(figsize=(10, 6))  # Adjust size as needed
sns.lineplot(x='episode', y='ewm', data=result_df_rgb_training, sort=False)
plt.xlabel('Episode')
plt.ylabel('Average Total Reward (EWM)')
plt.title('Average Total Reward per Episode (EWM)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout
plt.show()
```
