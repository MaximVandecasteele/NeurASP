```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# RGB baseline

```python
file_path = '/Users/maximvandecasteele/PycharmProjects/NeurASP/RL/training_run_baseline/tensorboard_rgb/csv/rgb_1.csv'
file_path2 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/RL/training_run_baseline/tensorboard_rgb/csv/rgb_2.csv'
file_path3 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/RL/training_run_baseline/tensorboard_rgb/csv/rgb_3.csv'
file_path4 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/RL/training_run_baseline/tensorboard_rgb/csv/rgb_4.csv'
file_path5 = '/Users/maximvandecasteele/PycharmProjects/NeurASP/RL/training_run_baseline/tensorboard_rgb/csv/rgb_5.csv'
```

```python
df1 = pd.read_csv(file_path).drop('Wall time', axis=1)
df2 = pd.read_csv(file_path2).drop('Wall time', axis=1)
df3 = pd.read_csv(file_path3).drop('Wall time', axis=1)
df4 = pd.read_csv(file_path4).drop('Wall time', axis=1)
df5 = pd.read_csv(file_path5).drop('Wall time', axis=1)
```

```python
df1['Step']
```

```python
value_columns = [df['Value'] for df in [df1, df2, df3, df4, df5]]
concatenated_values = pd.concat(value_columns, axis=1)
result_df = pd.DataFrame(concatenated_values.mean(axis=1), columns=['mean_value'])
result_df1 = pd.concat([df1['Step'], result_df], axis=1)
```

```python
result_df1
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Plot the average reward per model
plt.figure(figsize=(10, 6))  # Adjust size as needed
sns.lineplot(x='Step', y='mean_value', data=result_df1, sort=False)
plt.xlabel('Episode')
plt.ylabel('Average Total Reward')
plt.title('Average Total reward per Episode')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
result_df1['ewm'] = result_df1.ewm(alpha=0.1).mean()['mean_value']
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Plot the average reward per model
plt.figure(figsize=(10, 6))  # Adjust size as needed
sns.lineplot(x='Step', y='ewm', data=result_df1, sort=False)
plt.xlabel('Episode')
plt.ylabel('Average Total Reward (EWM)')
plt.title('Average Total Reward per Episode (EWM)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout
plt.show()
```
