```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
file_path = '/Users/maximvandecasteele/PycharmProjects/NeurASP/mario_vanilla/B1/eval/performance_B1.csv'
```

```python
df = pd.read_csv(file_path)
sorted_df = df.sort_values(by=['model', 'episode'], ascending=True)
```

```python
sorted_df
```

```python
result_df = sorted_df.groupby('model')['reward'].mean().reset_index()
```

```python
result_df
```

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Plot the average reward per model
plt.figure(figsize=(10, 6))  # Adjust size as needed
sns.lineplot(x='model', y='reward', data=result_df, marker='o', sort=False)
plt.xlabel('Model')
plt.ylabel('Average Reward')
plt.title('Average Reward per Model')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
sns.regplot(x='model', y='reward', data=result_df)
```
